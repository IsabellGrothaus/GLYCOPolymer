# Calculation of the persistence length

$\langle \vec{t}_{j,norm} \cdot \vec{t}_{i,norm} \rangle = \cos(\theta_i) = e^{-\frac{i}{L_p}}$

<br>
- Positions: $\vec{x}_{i}$<br>
- Tangent Vectors: $\vec{t}_{i} = \vec{x}_{i+1}-\vec{x}_{i}$<br>
- Norm Tangent Vectors: $\|\vec{t}_{i}\|$<br>
- Normalized Tangent Vectors: $\vec{t}_{i,norm} = \frac{\vec{t_{i}}}{\|\vec{t_{i}}\|}$<br>
- Cosine of angles: $\cos(\theta_{i})=\vec{t}_{j,norm} \cdot \vec{t}_{i,norm}$<br>
- Persistence Length = $L_p$

### Initialize lists for data storage:

```python
# Initialize emptys lists to store data over all frames
positions_all_frames = []
tangent_vectors_all_frames = []
norm_tangent_vectors_all_frames = []
normalized_tangent_vectors_all_frames = []
tangents_cosine_all_frames1 = []
tangents_cosine_all_frames2 = []
tangents_cosine_all_frames3 = []
```

### Calculate positions, vectors and angles for all time steps in the trajectory and store in arrays:

```python
# Iterate through the frames in the trajectory
for ts in u.trajectory:

    # Load positions of all O3 atoms of all residues
    positions = polymer_chain.positions
    
    # Calculate the tangent vectors for the current frame
    tangent_vectors = np.diff(positions, axis=0)
    
    # Calculate norm of each tangent vector
    norm_tangent_vectors = np.linalg.norm(tangent_vectors, axis=1)
    
    # Calculate normalized tangent vector
    normalized_tangent_vectors = (tangent_vectors.T / norm_tangent_vectors).T
    
    # Initialize empty local list for storing cosine of tangents
    tangents_cosine1 = []
    tangents_cosine2 = []
    tangents_cosine3 = []
    
    # Calculate each cosine between the tangents one by one for resid 1 to i-2, starting from tangent vector 0 to i-3
    for i in range(len(normalized_tangent_vectors)-3):
        s = np.dot(normalized_tangent_vectors[0],normalized_tangent_vectors[i]) 
        tangents_cosine1.append(s)
        
    # Calculate each cosine between the tangents one by one for residue 2 to i-1, starting from tangent vector 1 to i-2
    for i in range(len(normalized_tangent_vectors)-3):
        s = np.dot(normalized_tangent_vectors[1],normalized_tangent_vectors[i+1]) 
        tangents_cosine2.append(s)
        
    # Calculate each cosine between the tangents one by one residue 3 to i, starting from tangent vector 2 to i-1
    for i in range(len(normalized_tangent_vectors)-3):
        s = np.dot(normalized_tangent_vectors[2],normalized_tangent_vectors[i+2]) 
        tangents_cosine3.append(s)    

    # Append all positions
    positions_all_frames.append(positions)    

    # Append the tangent vectors 
    tangent_vectors_all_frames.append(tangent_vectors)
    
    # Append the norm of all tangents vectors 
    norm_tangent_vectors_all_frames.append(norm_tangent_vectors)
    
    # Append the normalized tangent vectors 
    normalized_tangent_vectors_all_frames.append(normalized_tangent_vectors)
    
    #Append the cosine of all tangents 
    tangents_cosine_all_frames1.append(tangents_cosine1)
    
    #Append the cosine of all tangents
    tangents_cosine_all_frames2.append(tangents_cosine2)
    
    #Append the cosine of all tangents 
    tangents_cosine_all_frames3.append(tangents_cosine3)
```

### Calculate time averages of tangent vector $\langle \cos(\theta_{i}) \rangle$ & $\langle \vec{t}_{i,norm} \rangle$

```python
# Calculate the average cosine of tangent vectors over all frames
average_tangents_cosine1 = np.mean(tangents_cosine_all_frames1, axis=0)
average_tangents_cosine2 = np.mean(tangents_cosine_all_frames2, axis=0)
average_tangents_cosine3 = np.mean(tangents_cosine_all_frames3, axis=0)

average_tangents_cosine = np.mean([average_tangents_cosine1, average_tangents_cosine2, average_tangents_cosine3], axis=0)

# Convert norm of all tangents vectors into numpy array to enable using np.mean and calculate time average t
np_norm_tangent_vectors_all_frames = np.array(norm_tangent_vectors_all_frames)
average_norm_tangent_vectors_all_frames = np.mean(np_norm_tangent_vectors_all_frames)
```

### Non-linear fitting of exponential decay function and calculation of persistence length: $l_{p}$

```python
# Define the exponential decay function for fitting
def exp_decay(i, Lp):
    return np.exp(-i / Lp)

# Prepare the x-axis data (number of tangent vectors)
tangents_i = np.arange(len(average_tangents_cosine))

# Perform the curve fitting
popt, pcov = curve_fit(exp_decay, tangents_i, average_tangents_cosine)

# Extract the persistence length from the fit parameters
persistence_length = popt[0]
persistence_length_angstroms = persistence_length*average_norm_tangent_vectors_all_frames
```

### Plotting $\langle \cos(\theta_{i}) \rangle$ over $i$ and  $l_{p}$

```python
# Plotting with fit
fig = plt.figure(figsize=(10, 8))
plt.plot(tangents_i, average_tangents_cosine, 'k-', label='$ <cos(θ_{i})>$')
plt.plot(tangents_i, exp_decay(tangents_i, *popt), 'r--', label=f'Curve fit')

plt.legend(edgecolor = 'black', fontsize = 25,loc='lower right')

box_text = '$L_p$ = ' + str(persistence_length.round(1)) + ' bonds, ' + str(persistence_length_angstroms.round(1)) + ' Å'
plt.text(5, 1, box_text, fontsize=25, ha='center', va='center', bbox=dict(boxstyle='round', facecolor='white', edgecolor='black', alpha=0.7))

plt.xlabel('Number of Tangent Vectors $r_{i}$', fontsize = 25)
plt.ylabel('$ <cos(θ_{i})>$', fontsize = 25)
plt.ylim(0, 1.1)  # Set y-axis limits
#plt.xlim(0, 103)  # Set y-axis limits
plt.xticks(fontsize= 25)
plt.yticks(fontsize= 25)

plt.tick_params(axis='y', which='both', labelbottom=True, labelleft=True)
plt.tick_params(axis='x', which='both', labelbottom=True, labelleft=True)

plt.tick_params(axis='y', which='both', length=5) 
plt.tick_params(axis='x', which='both', length=5) 

# Save the figure (Choose either png, svg ord pdf by commenting)
# plt.savefig('persistence_length_plot.png', dpi=1200)  # Save as PNG with high dpi
plt.savefig('Persistence_Length25_sulf.svg')  # Save as SVG
# plt.savefig('persistence_length_plot.pdf')  # Save as PDF

plt.show()

# Print the persistence length
print(f"Persistence Length: {persistence_length} in Bonds")
print(f"Persistence Length: {persistence_length_angstroms} in Å")
```

## Block average of persistence length

### Calculate persistence length for each slice

```python
# Total number of frames in the trajectory
total_frames = len(u.trajectory)

# Number of slices
num_slices = 5

# The size of each slice, rounded up to ensure all frames are included
slice_size = ceil(total_frames / num_slices)

# Define the exponential decay function for fitting
def exp_decay(i, Lp):
    return np.exp(-i / Lp)

# Arrays to store the persistence length of each slice and the slice indices
persistence_lengths = []
slice_indices = []

# Process each slice
for slice_num in range(num_slices):
    slice_start = slice_num * slice_size
    slice_end = min((slice_num + 1) * slice_size, total_frames)
    tangents_cosine_all_frames1 = []
    tangents_cosine_all_frames2 = []
    tangents_cosine_all_frames3 = []
            
    # Iterate through frames in the current slice
    for ts in u.trajectory[slice_start:slice_end]:
        positions = polymer_chain.positions
        tangent_vectors = np.diff(positions, axis=0)
        norm_tangent_vectors = np.linalg.norm(tangent_vectors, axis=1)
        normalized_tangent_vectors = (tangent_vectors.T / norm_tangent_vectors).T

        # Ensure that the array lengths are consistent
        num_vectors = len(normalized_tangent_vectors) - 3

        # Calculate cosine between tangents
        tangents_cosine1 = [np.dot(normalized_tangent_vectors[0], normalized_tangent_vectors[i]) for i in range(num_vectors)]
        tangents_cosine_all_frames1.append(tangents_cosine1)

        tangents_cosine2 = [np.dot(normalized_tangent_vectors[1], normalized_tangent_vectors[i+1]) for i in range(num_vectors)]
        tangents_cosine_all_frames2.append(tangents_cosine2)

        tangents_cosine3 = [np.dot(normalized_tangent_vectors[2], normalized_tangent_vectors[i+2]) for i in range(num_vectors)]
        tangents_cosine_all_frames3.append(tangents_cosine3)

    # Calculate the average cosine of tangent vectors over all frames in the slice
    average_tangents_cosine1 = np.mean(tangents_cosine_all_frames1, axis=0)
    average_tangents_cosine2 = np.mean(tangents_cosine_all_frames2, axis=0)
    average_tangents_cosine3 = np.mean(tangents_cosine_all_frames3, axis=0)
    
    # Calculate the average of the before defined tangent cosines
    average_tangents_cosine_all = np.mean([average_tangents_cosine1, average_tangents_cosine2, average_tangents_cosine3], axis=0)
    average_norm_tangent_vectors = np.mean(norm_tangent_vectors)

    # Perform the curve fitting
    tangents_i = np.arange(len(average_tangents_cosine_all))
    popt, _ = curve_fit(exp_decay, tangents_i, average_tangents_cosine_all)
    persistence_length = popt[0] * average_norm_tangent_vectors
    persistence_lengths.append(persistence_length)
    slice_indices.append(slice_num + 1)
```

### Calculate and print average persistence length 

```python
# Calculate the average and standard deviation of the persistence lengths
average_persistence_length = np.mean(persistence_lengths)
std_persistence_length = np.std(persistence_lengths)

print(f"Average Persistence Length: {average_persistence_length} in Å")
print(f"Standard Deviation of Persistence Lengths: {std_persistence_length} in Å")
```

# Calculation of the radius of gyration

$ R_g = \sqrt{\frac{1}{N}\sum_{i=1}^{N}(\vec{x}_i - \vec{x}_{cm})^2} $

- $ R_g $: Radius of gyration.
- $ N $: Number of backbone residue units in the polymer chain.
- $ x_i $: Position of the \( i \)-th backbone residue unit.
- $ x_{cm}$: Position of the center of mass of the polymer chain.

## Block Averaging

### Calculate radius of gyration for each slice

```python
# Total number of frames in the trajectory
total_frames = len(u.trajectory)

# Number of slices
num_slices = 5

# The size of each slice, rounded up
slice_size = ceil(total_frames / num_slices)

# Lists to store the average radius of gyration for each slice and the slice indices
radius_gyration_per_slice = []
slice_indices = []

# Process each slice
for slice_num in range(num_slices):
    slice_start = slice_num * slice_size
    slice_end = min((slice_num + 1) * slice_size, total_frames)
    radius_gyration_values = []
    
    # Iterate through frames in the current slice
    for ts in u.trajectory[slice_start:slice_end]:
        center_of_mass = polymer_chain.center_of_mass()
        distances = polymer_chain.positions - center_of_mass
        squared_distances = np.mean(np.sum(distances**2, axis=1))
        radius_gyration = np.sqrt(squared_distances)
        radius_gyration_values.append(radius_gyration)
    
    # Calculate the average radius of gyration for the slice
    average_radius_gyration = np.mean(radius_gyration_values)
    radius_gyration_per_slice.append(average_radius_gyration)
    slice_indices.append(slice_num + 1)
    
```

```python
# Calculate the overall average and standard deviation
overall_average_radius_gyration = np.mean(radius_gyration_per_slice)
overall_std_radius_gyration = np.std(radius_gyration_per_slice)
```

## Calculate positions, vectors and angles for all time steps in the trajectory and store in arrays:
- Positions: $\vec{x}_{i}$<br>
- Center of Mass: $\vec{x}_{cm}$ via MDAnalysis.Universe.center_of_mass()

```python
# Array to store radii of gyration for each frame
radius_gyration_values = np.zeros(len(u.trajectory))

# Calculate radius of gyration for each frame
for ts in u.trajectory:
    
    # Calculate the center of mass
    center_of_mass = polymer_chain.center_of_mass()
    
    # Calculate distances between each residues' C1 and the center of mass
    distances = polymer_chain.positions - center_of_mass
    
    # Square the distances
    squared_distances = np.mean(np.sum(distances**2, axis=1))
    
    # Calculate radius of gyration via Square Root of squared distances
    radius_gyration_values[ts.frame] = np.sqrt(squared_distances)

# Calculate the average radius of gyration
average_radius_gyration = np.mean(radius_gyration_values)

# Load time step size from Universe
time_per_frame_ps = u.trajectory.dt

# Calculate time for each frame using dt
times_ns = np.arange(0, len(u.trajectory)) * time_per_frame_ps / 1000
```

### Plotting $\langle R_{g} \rangle$ over $t$

```python
# Plot the radius of gyration over time in ns
plt.figure(figsize=(10, 8))
plt.scatter(times_ns[::5], radius_gyration_values[::5], color='black', s=3, label='$R_{g}$')
plt.axhline(overall_average_radius_gyration, color='red', linestyle='--', linewidth=3, label='$\langle R_{g}⟩$')

plt.legend(edgecolor = 'black', fontsize = 25,loc='lower left')

box_text = '$ \langle {R_G}⟩$ = ' + str(overall_average_radius_gyration.round(1)) + '±' + str(overall_std_radius_gyration.round(1)) + ' Å'  # Change this to your desired number
plt.text(750, 15.2, box_text, fontsize=25, ha='center', va='center', bbox=dict(boxstyle='round', facecolor='white', edgecolor='black', alpha=0.7))

plt.xlabel('Time (ns)', fontsize = 25)
plt.ylabel('Radius of Gyration (Å)', fontsize = 25)
#plt.ylim(0, 130)  # Set y-axis limits
plt.xticks(fontsize= 25)
plt.yticks(fontsize= 25)

plt.tick_params(axis='y', which='both', labelbottom=True, labelleft=True)
plt.tick_params(axis='x', which='both', labelbottom=True, labelleft=True)

plt.tick_params(axis='y', which='both', length=5) 
plt.tick_params(axis='x', which='both', length=5) 

plt.show()

# Save the figure (Choose either png, svg ord pdf by commenting)
# plt.savefig('persistence_length_plot.png', dpi=1200)  # Save as PNG with high dpi
plt.savefig('Radius_of_gyration25_sulf.svg')  # Save as SVG
# plt.savefig('persistence_length_plot.pdf')  # Save as PDF

print(f"Overall average radius of gyration: {overall_average_radius_gyration} Å")
print(f"Standard deviation of radii of gyration: {overall_std_radius_gyration} Å")
```

# Calculation of the contour length and the end-to-end vector

$\mathbf{R} = \mathbf{x}_{i=n} - \mathbf{x}_1$

$\mathbf{R}$: Vector connecting the positions of the first and last backbone residue (or atoms) in a chain. <br>
$\mathbf{x}_1$: Position vector of the first backbone residue<br>
$\mathbf{x}_{i=n}$: Position vector of the last backbone residue in a chain of $n$ backbone residues

### Block Average, Calculating End-to-End vector for each slice 

```python
# Total number of frames in the trajectory
total_frames = len(u.trajectory)

# Number of slices
num_slices = 5

# The size of each slice, rounded up
slice_size = ceil(total_frames / num_slices)

# Arrays to store the average norm of the end-to-end vector and contour length for each slice, and the slice indices
average_end_to_end_per_slice = []
contour_length_per_slice = []
norm_tangent_per_slice = []
slice_indices = []


# Process each slice
for slice_num in range(num_slices):
    slice_start = slice_num * slice_size
    slice_end = min((slice_num + 1) * slice_size, total_frames)
    norm_end_to_end_vectors = []
    contour_lengths = []
    norm_tangent_vectors = []  # Move this line inside the loop
    
    # Iterate through frames in the current slice
    for ts in u.trajectory[slice_start:slice_end]:
        positions = polymer_chain.positions
        
        # Calculate the end-to-end vector for the current frame
        end_to_end_vector = positions[-1] - positions[0]
        
        # Calculate the norm of the end-to-end vector
        norm_end_to_end_vector = np.linalg.norm(end_to_end_vector)
        norm_end_to_end_vectors.append(norm_end_to_end_vector)
        
        # Calculate the tangent vectors for the current frame
        tangent_vectors = np.diff(positions, axis=0)
        
        # Calculate the norm of each tangent vector
        norm_tangent_vector = np.linalg.norm(tangent_vectors, axis=1)
        norm_tangent_vectors.extend(norm_tangent_vector)
        
        # Calculate the contour length for the current frame
        contour_length = np.sum(norm_tangent_vector)
        contour_lengths.append(contour_length)
    
    # Calculate the average norm of the end-to-end vector, contour length, 
    # and average norm of tangent vectors for the slice
    average_norm_end_to_end = np.mean(norm_end_to_end_vectors)
    average_end_to_end_per_slice.append(average_norm_end_to_end)
    
    average_contour_length = np.mean(contour_lengths)
    contour_length_per_slice.append(average_contour_length)
    
    average_norm_tangent_vectors = np.mean(norm_tangent_vectors)
    norm_tangent_per_slice.append(average_norm_tangent_vectors)
    
    slice_indices.append(slice_num + 1)
```

### Calculating average and standard deviation of end-to-end vector, contour length and bond length for block averaging

```python
# Calculate the overall average and standard deviation
overall_average_end_to_end = np.mean(average_end_to_end_per_slice)
overall_std_end_to_end = np.std(average_end_to_end_per_slice)

overall_average_contour_length = np.mean(contour_length_per_slice)
overall_std_contour_length = np.std(contour_length_per_slice)

overall_average_norm_tangent_vectors = np.mean(norm_tangent_per_slice)
overall_std_norm_tangent_vectors = np.std(norm_tangent_per_slice)


print(f"Average Norm of Tangent Vectors: {overall_average_norm_tangent_vectors}")
print(f"Standard Deviation of Norm of Tangent Vectors: {overall_std_norm_tangent_vectors}")
```

### Calculate end-to-end vectors and contour length for all time steps in the trajectory and store in arrays:
<br>
- Positions: $\vec{x}_{i}$<br>
- Tangent Vectors: $\vec{t}_{i} = \vec{x}_{i+1}-\vec{x}_{i}$<br>
- Norm Tangent Vectors: $\|\vec{r}_{i}\|$<br>
- End-to-end-vector: $\vec{R} = \vec{x}_{i=n} - \vec{x}_{i=1}$ with $n$ denoting the last C1 atom in the topology<br>
- Contour length: $\| L \|$

#### Initialize arrays for data storage:

```python
# Initialize emptys lists to store data over all frames
end_to_end_vector_all_frames = []
norm_end_to_end_vector_all_frames = []
contour_length_all_frames = []
tangent_vectors_all_frames = []
norm_tangent_vectors_all_frames = []
normalized_tangent_vectors_all_frames = []
tangents_cosine_all_frames = []
positions_all_frames = []
```

#### Calculation over all frames

```python
# Iterate through the frames in the trajectory
for ts in u.trajectory:

    # Load positions of all C1 Atoms of all residues
    positions = polymer_chain.positions
    
    #Calculate End to End Vector
    end_to_end_vector = positions[-1] - positions[0]
    
    # Calculate the Norm (Betrag) of the end_to_end_vector to get absolute distance in Angstroms
    norm_end_to_end_vector = np.linalg.norm(end_to_end_vector)
    
    # Calculate the tangent vectors for the current frame
    tangent_vectors = np.diff(positions, axis=0)
    
    # Calculate norm of each tangent vector
    norm_tangent_vectors = np.linalg.norm(tangent_vectors, axis=1)
    
    # Calculate the contour length via the sum of all norms of all tangent vectors
    contour_length = np.sum(norm_tangent_vectors)
    
    
    #Append all positions
    positions_all_frames.append(positions)   

    #Append end to end vector to the array 
    end_to_end_vector_all_frames.append(end_to_end_vector)    
    
    # Append the length of the end to end vector to the array
    norm_end_to_end_vector_all_frames.append(norm_end_to_end_vector)

    # Append the tangent vectors to the array
    tangent_vectors_all_frames.append(tangent_vectors)
    
    # Append the norm of all tangents vectors to the array
    norm_tangent_vectors_all_frames.append(norm_tangent_vectors)
    
    # Append all contour lengths of all ts for debugging purposes
    contour_length_all_frames.append(contour_length)
```

#### Calculate time array in ns

```python
# Load time step size from Universe
time_per_frame_ps = u.trajectory.dt

# Calculate time for each frame using dt
times_ns = np.arange(0, len(u.trajectory)) * time_per_frame_ps / 1000
```

#### Plotting $\langle \| \vec{R} \| \rangle$ over $t$ 

```python
# Plot the Norm of the end-to-end-vector over time in ns and the time average of the norm of end-to-end-vector and contour length
fig = plt.figure(figsize=(10, 8))
plt.scatter(times_ns[::5], norm_end_to_end_vector_all_frames[::5], color='black', s=3, label='${|R|}$')
plt.axhline(overall_average_end_to_end, color='red', linestyle='--', linewidth=3, label='$\langle R⟩$')
plt.axhline(overall_average_contour_length, color='green', linestyle='--', linewidth=3, label='$\langle L⟩$')

plt.legend(edgecolor = 'black', fontsize = 25, loc= 'upper right')

box_text = '$ \langle {R}⟩$ = ' + str(overall_average_end_to_end.round(1)) + '±' + str(overall_std_end_to_end.round(1)) + ' Å' + '\n' + '$ \langle {L}⟩$ = ' + str(overall_average_contour_length.round(1)) + '±' + str(overall_std_contour_length.round(1)) + ' Å' # Change this to your desired number
plt.text(300, 50, box_text, fontsize=25, ha='center', va='center', bbox=dict(boxstyle='round', facecolor='white', edgecolor='black', alpha=0.7))

plt.xlabel('Time (ns)', fontsize = 25)
plt.ylabel('Norm of the end-to-end-vector (Å)', fontsize = 25)
#plt.ylim(0, 450)  # Set y-axis limits
plt.xticks(fontsize= 25)
plt.yticks(fontsize= 25)

plt.tick_params(axis='y', which='both', labelbottom=True, labelleft=True)
plt.tick_params(axis='x', which='both', labelbottom=True, labelleft=True)

plt.tick_params(axis='y', which='both', length=5) 
plt.tick_params(axis='x', which='both', length=5) 

# Save the figure (Choose either png, svg ord pdf by commenting)
# plt.savefig('persistence_length_plot.png', dpi=1200)  # Save as PNG with high dpi
plt.savefig('Persistence_Length25_sulf.svg')  # Save as SVG
# plt.savefig('persistence_length_plot.pdf')  # Save as PDF

plt.show()


# Print average end-to-end vector
print(f"Overall Average Norm of End-to-End Vector: {overall_average_end_to_end} Å")
print(f"Standard Deviation of End-to-End Vectors: {overall_std_end_to_end} Å")

# Print average contour length 
print(f"Average Contour Length: {overall_average_contour_length}")
print(f"Standard Deviation of Contour Length: {overall_std_contour_length}")

# Print average bond length 
print(f"Average Norm of Tangent Vectors: {overall_average_norm_tangent_vectors}")
print(f"Standard Deviation of Norm of Tangent Vectors: {overall_std_norm_tangent_vectors}")
```
